```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from functools import reduce

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# load in csv files
# enrollment data
enr_14_load = os.path.join('Enrollment Report_District_AUSTIN ISD_Campuses_Ethnicity_2013-2014.csv')
enr_15_load = os.path.join('Enrollment Report_District_AUSTIN ISD_Campuses_Ethnicity_2014-2015.csv')
enr_16_load = os.path.join('Enrollment Report_District_AUSTIN ISD_Campuses_Ethnicity_2015-2016.csv')
enr_17_load = os.path.join('Enrollment Report_District_AUSTIN ISD_Campuses_Ethnicity_2016-2017.csv')
enr_18_load = os.path.join('Enrollment Report_District_AUSTIN ISD_Campuses_Ethnicity_2017-2018 .csv')

# student data
std_15_load = os.path.join('StudPgmDistrictCampus15district227901.csv')
std_16_load = os.path.join('StudPgmDistrictCampus16district227901.csv')
std_17_load = os.path.join('StudPgmDistrictCampus17district227901.csv')
std_18_load = os.path.join('StudPgmDistrictCampus18district227901.csv')

```

```python
# transfer data into dataframes
# enrollment data
enr_14 = pd.read_csv(enr_14_load, skiprows=2)
enr_15 = pd.read_csv(enr_15_load, skiprows=2)
enr_16 = pd.read_csv(enr_16_load, skiprows=2)
enr_17 = pd.read_csv(enr_17_load, skiprows=2)
enr_18 = pd.read_csv(enr_18_load, skiprows=2)

# student data
std_15 = pd.read_csv(std_15_load)
std_16 = pd.read_csv(std_16_load)
std_17 = pd.read_csv(std_17_load)
std_18 = pd.read_csv(std_18_load)
```

```python
# create function to clean data

# put all dataframes into a list
data_frames = [std_15, std_16, std_17, std_18]

# create a function to clean all dataframes
def func(df):
    df2 = df.replace(-9999999, 2)
    df3 = df2[['CAMPUS NAME', 'TOTAL ENROLLMENT', 'TOTAL ECONOMICALLY DISADVANTAGED STUDENTS', 
              'TOTAL TITLE I STUDENTS', 'TOTAL AT RISK STUDENTS']]
#     df4 = df3.groupby('CAMPUS NAME').sum().reset_index()
    return df3
    
```

```python
# call function
for i in range(0,len(data_frames)):
    data_frames[i] = func(data_frames[i])
```
